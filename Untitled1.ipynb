```python
import pandas as pd
import json

# Load the user's dataset
with open('job_salary_converted.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)

print("--- Step 1: Raw Data (First 5 rows) ---")
print(df.head())
```

```python
# Removing duplicate rows if any exist
df = df.drop_duplicates()

print(f"\n--- Step 2: Duplicates Removed ---")
print(f"Remaining rows: {len(df)}")
```

```python
# Fill missing job titles with 'Unknown' and missing years with the Mean
df['job_title'] = df['job_title'].fillna('Unknown')
df['experience_years'] = df['experience_years'].fillna(df['experience_years'].mean())

print("\n--- Step 3: Missing Values Handled ---")
```

```python
# Force Salary to be numeric; non-numeric values become NaN
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# Fill any new NaNs with the median salary
df['salary'] = df['salary'].fillna(df['salary'].median())

print("\n--- Step 4: Salary Cast to Numeric ---")
```

```python
# If experience_years is abnormally high (e.g., > 50), set it to the average (e.g., 10)
df.loc[df['experience_years'] > 50, 'experience_years'] = 10

# Standardise job titles to Title Case (e.g., 'ai engineer' -> 'AI Engineer')
df['job_title'] = df['job_title'].str.title()

print("\n--- Step 5: Outliers Fixed & Text Standardised ---")
```

```python
# Ensure remote_work follows a standard format if it were mixed
# (Example: Converting strings to categorical or standardising case)
df['remote_work'] = df['remote_work'].str.strip().str.title()

print("\n--- Step 6: Final Cleaned Data (First 5 rows) ---")
print(df.head())
```
